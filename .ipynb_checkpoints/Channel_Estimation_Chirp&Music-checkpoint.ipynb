```python
import numpy as np
import matplotlib.pyplot as plt
import pyaudio
import wave
from pydub import AudioSegment
import scipy
from scipy.io import wavfile #audio i/o
from IPython.display import Audio #listening to audio
import thinkdsp
import thinkstats2
%matplotlib inline
```

```python
signal = thinkdsp.Chirp(start = 1, end = 22050, amp = 1.0)
wave = signal.make_wave(duration = 10, framerate = 44100)
wave.plot()
spectrum = wave.make_spectrum()
plt.figure()
spectrum.plot()
wave.make_audio()
```

```python
music = thinkdsp.read_wave('music.wav').segment(start = 0, duration = 10)
music.plot()
music.make_audio()
```

```python
recv = thinkdsp.read_wave('Chirp_music_20cm.wav')
chirp_init = 1
music_init = 14
recv.plot()
recv_chirp = recv.segment(start = chirp_init, duration = 12)
plt.figure()
recv_chirp.plot()
recv_music = recv.segment(start = music_init, duration = 12)
plt.figure()
recv_music.plot()
```

First estimate the channel through the Chirp

```python
corr = np.correlate(recv_chirp.ys, wave.ys, 'valid')
start_chirp = np.argmax(corr)
start_chirp
```

```python
recv_chirp_seg = recv.segment(start = chirp_init+(start_chirp-50)/44100, duration = 10.01)
recv_chirp_seg.plot()
```

```python
H_chirp = np.fft.fft(recv_chirp_seg.ys)/np.fft.fft(wave.ys, len(recv_chirp_seg.ys))
cir_chirp = np.fft.ifft(H_chirp)
x = np.linspace(0, len(cir_chirp)/44100, len(cir_chirp))
plt.plot(x,cir_chirp)
```

```python
x = np.linspace(0, len(cir_chirp[:300])/44100, len(cir_chirp[:300]))
plt.plot(x,cir_chirp[:300])
```

```python
corr = np.correlate(recv_music.ys, music.ys, 'valid')
start_music = np.argmax(corr)
start_music
```

```python
recv_music_seg = recv.segment(start = music_init+(start_music-150)/44100, duration = 10.01)
recv_music_seg.plot()
```

```python
H_music = np.fft.fft(recv_music_seg.ys)/np.fft.fft(music.ys, len(recv_music_seg.ys))
cir_music = np.fft.ifft(H_music)
x = np.linspace(0, len(cir_music)/44100, len(cir_music))
plt.plot(x,cir_music)
```

```python
x = np.linspace(0, len(cir_music[:300])/44100, len(cir_music[:300]))
plt.plot(x,cir_music[:300])
```

```python
plt.figure(figsize = (20,6))
plt.subplot(121)
x = np.linspace(0, len(cir_chirp[:300])/44100, len(cir_chirp[:300]))
plt.plot(x,cir_chirp[:300])
plt.title('Chirp')
plt.subplot(122)
x = np.linspace(0, len(cir_music[:300])/44100, len(cir_music[:300]))
plt.plot(x,cir_music[:300])
plt.title('Music')
```

```python
chirp_ch_rfft = np.fft.rfft(recv_chirp_seg.ys)/np.fft.rfft(wave.ys, len(recv_chirp_seg.ys))
music_ch_rfft = np.fft.rfft(recv_music_seg.ys)/np.fft.rfft(music.ys, len(recv_music_seg.ys))

plt.figure(figsize = (20,6))
plt.subplot(121)
x = np.linspace(0, 44100/2, len(music_ch_rfft))
plt.plot(x, chirp_ch_rfft)
plt.title('Chirp')
plt.subplot(122)
plt.plot(x, music_ch_rfft)
plt.title('Music')
```

```python
peak_chirp = np.argmax(cir_chirp)
peak_music = np.argmax(cir_music)
print(peak_chirp)
print(peak_music)
```
